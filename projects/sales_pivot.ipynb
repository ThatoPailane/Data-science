```python
import pandas as pd
```

```python
def pivot(file_path, result_path):

    df = pd.read_excel(file_path)

    sales_table = df.pivot_table(
        index='Category',
        columns='Region',
        values='Sales',
        aggfunc='sum',
        fill_value=0,
    )

    quantity_table = df.pivot_table(
        index='Category',
        columns='Region',
        values='Quantity',
        aggfunc='mean',
        fill_value=0
    )

    sales_table = sales_table.astype(float)
    
    combined_table = pd.merge(sales_table, quantity_table)

    final_table = combined_table.reset_index(drop=True)

    final_table.to_excel(result_path)
```

```python
pivot("assets/Sales_Data.xlsx","assets/pivot_result.xlsx")
```
